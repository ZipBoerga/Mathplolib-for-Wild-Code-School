# Execute the code below

```
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import datetime as dt

link = "https://raw.githubusercontent.com/murpi/wilddata/master/quests/weather2019.csv"
df_weather = pd.read_csv(link).iloc[:,[0, 1, 2, -3]]

df_weather['DATE'] = pd.to_datetime(df_weather['DATE'])
```

# Now, complete the following tasks :

## ScatterPlot with 2 variables, title and legend
First of all, you have to create a scatterplot with MAX (in red) and MIN (in blue) Temperatures on the same graph

X-axis stay "DATE" for both values

You have to add title and legend

Here is an example of what is expected :

matplotlib1.png

```
plt.scatter(x = df_weather["DATE"], y = df_weather["MAX_TEMPERATURE_C"], c= 'green', label='Max')
plt.scatter(x = df_weather["DATE"], y = df_weather["MIN_TEMPERATURE_C"], c= 'yellow', label='Min')
plt.legend(title='Temperatures')
plt.ylabel('Temperature')
plt.title('MIN and MAX temperatures')
plt.show()
```

## 2 charts on the same figure
Now, complete this code to have MAX and MIN Temperature in 2 charts, MAX on the left side and MIN on the right side

The X axis have to be the date of the year

Each graph must have a title, a legend for the colors, and of course, readable axis (ticks)

Here is an example of what is expected :


Capture d’écran 2021-08-26 à 17.27.31.png

```
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))

ax1 = plt.subplot(121)
ax1.scatter(df_weather["DATE"], df_weather["MAX_TEMPERATURE_C"], c='green', label='Max')
ax1.set_ylabel('Temperature')
ax1.set_title('MAX Temperature')

ax2 = plt.subplot(122)
ax2.scatter(df_weather["DATE"], df_weather["MIN_TEMPERATURE_C"], c='yellow', label='Min')
ax2.set_ylabel('Temperature')
ax2.set_title('MIN Temperature')

y_min = min(df_weather["MIN_TEMPERATURE_C"] - 5)
y_max = max(df_weather["MAX_TEMPERATURE_C"] + 5)
ax1.set_ylim(y_min, y_max)
ax2.set_ylim(y_min, y_max)

plt.show()
```
